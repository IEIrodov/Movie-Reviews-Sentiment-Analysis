# Data Cleaning


```python
documents = []
for category in movie_reviews.categories():
  for fileid in movie_reviews.fileids(category):
    documents.append((movie_reviews.words(fileid),category))

documents[0:5]
```

```python
import random
random.shuffle(documents)
documents[0:5]
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
#converting pos tags to simple pos tags which can be understood by Lemmatizer
from nltk.corpus import wordnet
def get_simple_pos(tag):
  if tag.startswith('J'):
    return wordnet.ADJ
  elif tag.startswith('V'):
    return wordnet.VERB
  elif tag.startswith('N'):
    return wordnet.NOUN
  elif tag.startswith('R'):
    return wordnet.ADV
  else:
    return wordnet.NOUN
```

```python
from nltk import pos_tag
w = "better"
pos_tag([w]) #we are supposed to pass pos_tag as an array not just a simple word
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
stops
```

```python
def clean_review(words):
  output_words = []
  for w in words:
    if w.lower() not in stops:
      pos = pos_tag([w])
      clean_word = lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
      output_words.append(clean_word.lower())
  return output_words
```

```python
documents = [(clean_review(document),category)for document,category in documents]
```

```python
training_documents = documents[0:1500]
testing_documents = documents[1500:]
```

```python
all_words = []
for doc in training_documents:
  all_words += doc[0]
```

```python
import nltk
```

```python
freq = nltk.FreqDist(all_words)
common = freq.most_common(3000)
features = [i[0] for i in common]
```

```python
features
```

```python
def get_feature_dict(words):
  current_features = {}
  words_set = set(words)
  for w in features:
    current_features[w] = w in words_set
  return current_features
```

```python
training_data = [(get_feature_dict(doc),category)for doc,category in training_documents]
testing_data = [(get_feature_dict(doc),category)for doc,category in testing_documents]
```

# Training Model

```python
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier,testing_data)
```

```python
classifier.show_most_informative_features(15)
```

# Using Scikit Learn in NLTK

```python
from sklearn.svm import SVC
from nltk.classify.scikitlearn import SklearnClassifier
```

```python
svc = SVC()
classifier_sklearn = SklearnClassifier(svc)
```

```python
classifier_sklearn.train(training_data)
```

```python
nltk.classify.accuracy(classifier_sklearn,testing_data)
```
